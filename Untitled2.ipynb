```python
file = open('words.txt', 'r')
text = []
for line in file.readlines():
    text = line.split()
global text
def cal_freq(text):
    stopwords = ['the', 'is', 'a', 'in', 'are', 'an', 'as', 'it', 'its', 'of', 'on', 'but', 'by', 'with', 'for' 
                ]
    puncs = ['.', ',', '!',';', ':']
    string = " ".join(text)
    new = []
    for i in string:
        if i not in puncs:
            new.append(i)
    string2 = "".join(new)
    contents = string2.lower().split()
    
    frequencies = {}
    for word in contents:
        if word not in frequencies:
            frequencies[word] = 1
        else:
            frequencies[word] +=1
    
    cloud = wordcloud.WordCloud()
    cloud.generate_from_frequencies(frequencies)
    return cloud.to_array()
```

```python
#display your cloud image

myimage = cal_freq(text)
plt.imshow(myimage, interpolation = 'nearest')
plt.axis('off')
plt.show()
```
